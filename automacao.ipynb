## Buscar dados da Ação

Bibliotecas: 
- yfinance (dados financeiros provenientes do Yahoo Finantial}
- matplotlib (Gráficos)

```python
# !pip install yfinance

import yfinance
```

```python
ticker = input("Informe a ação de sejada. Ex. PETR4.SA:")

dados = yfinance.Ticker(ticker).history(start="2020-01-01", end="2020-12-31") # Captura o histórico de uma ação direto da base do yahoo finantial
fechamento = (dados.Close) # Nome da coluna que se quer usar
fechamento.plot() #Aqui já usa a finção plot do matplotlib
```
